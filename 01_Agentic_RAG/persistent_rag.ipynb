# Persistent RAG

```python
from sqlitesearch import TextSearchIndex

sql_index = TextSearchIndex(
        text_fields=['question', 'section', 'answer'],
        keyword_fields=['course'],
        db_path='faq.db'
)
```

```python
# Check the number of records in the db
sql_index.count()
```

#### Try a search

```python
# Search the question
question = "Do I need to know GCP to join the course?"
db_results = sql_index.search(question,
            filter_dict = {'course': 'llm-zoomcamp'})
```

```python
# Check the number of results returned by sqlite
len(db_results)
```

```python
# Check the quality of questions returned by sqlite
for result in db_results:
    print(result['question'])
```
